```python
import sys
from caption import *
import urllib, cStringIO
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img_path='Images/test.png'
```

```python
model = Caption_Generator(mode = 'test')
```

```python
def get_image(path):
    file = cStringIO.StringIO(urllib.urlopen(path).read())
    img = Image.open(file)
    img.save(img_path)
    plt.imshow(img)
    plt.show()
    return img_path
```

```python
image_features, IDs = model.build_decode_graph()
```

```python
model.decode(image_features, IDs, get_image("Images/bike.jpg"))
```
